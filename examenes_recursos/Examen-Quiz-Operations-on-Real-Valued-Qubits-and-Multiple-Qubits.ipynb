```python
from math import pi
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

q = QuantumRegister(1) # quantum register with a single qubit
c = ClassicalRegister(1) # classical register with a single bit
qc = QuantumCircuit(q,c) # quantum circuit with quantum and classical registers
# Your code here
#qc.ry(4*pi/3,q[0])
qc.ry(4*pi/3,q[0])
# measure the qubit
qc.measure(q,c)
```

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# 1. Inicialización del Registro Cuántico, Clásico y el Circuito
q2 = QuantumRegister(2, "qreg")
c2 = ClassicalRegister(2, "creg")
qc2 = QuantumCircuit(q2, c2)

# 2. Preparación del estado entrelazado Phi+ (|00> + |11>)/sqrt(2)
qc2.h(q2[0])
qc2.cx(q2[0], q2[1])

# --- SOLUCIÓN: REEMPLAZO DE "# Your code here" ---
# Aplicamos una compuerta NOT (X) al segundo qubit para cruzar los estados.
# Esto convierte el término |00> en |01> y el término |11> en |10>.
qc2.x(q2[1])
# --------------------------------------------------

# 3. Medición de los registros cuánticos en los clásicos
qc2.measure(q2, c2)

# 4. Ejecución de la simulación con 1000 disparos (shots)
job = AerSimulator().run(qc2, shots=1000)
counts = job.result().get_counts(qc2)

# 5. Imprimir los resultados estadísticos obtenidos
print(counts)
```
