```python
from math import sqrt, isclose
import numpy as np

def _to_list(x):
    if isinstance(x, np.ndarray):
        return x.tolist()
    return x

def _igual(a, b, tol=1e-9):
    a = _to_list(a)
    b = _to_list(b)
    if isinstance(a, float) or isinstance(b, float):
        try:
            return isclose(float(a), float(b), rel_tol=tol, abs_tol=tol)
        except Exception:
            return False
    if isinstance(a, (list, tuple)) and isinstance(b, (list, tuple)):
        return len(a) == len(b) and all(_igual(x, y, tol) for x, y in zip(a, b))
    return a == b

def qprep_check(nombre, obtenido, esperado, tol=1e-9):
    if obtenido is None:
        print(f"{nombre}: pendiente. Reemplaza None por tu respuesta.")
        return False
    if _igual(obtenido, esperado, tol):
        print(f"{nombre}: correcto.")
        return True
    print(f"{nombre}: revisa tu resultado. Obtenido = {obtenido!r}")
    return False
```

## 1. Vectores como listas y como arreglos

Un vector real puede escribirse como una lista ordenada de componentes:

$$v = [v_1, v_2, \ldots, v_n].$$

En Python puro, una lista conserva el orden, pero no sabe que quieres hacer álgebra lineal. Para operaciones vectoriales usaremos dos enfoques:

- Funciones manuales, útiles para entender.
- NumPy, útil para calcular de forma compacta.

```python
v_lista = [1, 3, 5]
u_lista = [2, 4, 6]

v_np = np.array(v_lista)
u_np = np.array(u_lista)

print("Lista + lista:", v_lista + u_lista)     # concatenación
print("Array + array:", v_np + u_np)          # suma componente a componente
```

## 2. Operaciones vectoriales básicas

Para vectores de igual dimensión:

- Suma: $v + u = [v_1 + u_1, \ldots, v_n + u_n]$.
- Multiplicación por escalar: $c v = [c v_1, \ldots, c v_n]$.
- Longitud euclidiana: $\lVert v \rVert = \sqrt{v_1^2 + \cdots + v_n^2}$.

```python
def sumar_vectores(v, u):
    return [v[i] + u[i] for i in range(len(v))]

def escalar_vector(c, v):
    return [c * x for x in v]

def longitud(v):
    return sqrt(sum(x*x for x in v))

v = [1, 3, 5]
u = [2, 4, 6]

print("v + u =", sumar_vectores(v, u))
print("3v    =", escalar_vector(3, v))
print("||v|| =", longitud(v))
```

## 3. Longitud o norma euclidiana

Para $v=[1,3,5]$:

$$\lVert v\rVert = \sqrt{1^2 + 3^2 + 5^2} = \sqrt{35}.$$

En el quiz es usual que se pida redondear a dos dígitos después del punto decimal.

```python
v = [1, 3, 5]
print("suma de cuadrados =", sum(x*x for x in v))
print("longitud =", longitud(v))
print("redondeada a 2 decimales =", round(longitud(v), 2))
```

## 4. Producto punto (Yair)

Para dos vectores de igual dimensión:

$$v \cdot u = v^T u = \sum_{i=1}^{n} v_i u_i.$$

Ejemplo:

$$[1,2,3]\cdot[4,5,6] = 1\cdot4 + 2\cdot5 + 3\cdot6 = 32.$$

```python
def producto_punto(v, u):
    return sum(v[i] * u[i] for i in range(len(v)))

v = [1, 2, 3]
u = [4, 5, 6]

print(producto_punto(v, u))
print(np.dot(np.array(v), np.array(u)))
```

## 5. Ortogonalidad

Dos vectores reales no nulos son ortogonales si su producto punto es cero. En QBronze esto prepara la intuición para estados distinguibles, bases y mediciones.

```python
v = [1, 2, 3]
u = [-3, 6, -3]

print("v dot u =", producto_punto(v, u))
print("¿ortogonales?", producto_punto(v, u) == 0)
```

## 6. Notación $u^T u$, $u u^T$ y escalamiento

La expresión $u^T u$ es el producto punto de $u$ consigo mismo; produce un escalar:

$$u^T u = \lVert u \rVert^2.$$

Si $u$ se trata como vector columna, $u u^T$ produce una matriz. En ejercicios introductorios también puede aparecer una versión escalar-vector, por ejemplo $u_1 u^T$, donde $u_1$ es un escalar.

```python
u = np.array([1, 2, 2, 1])

u1 = u @ u              # u^T u
v = u1 * u              # u1 u^T visto como escalar por vector
v1 = v / 10

print("u^T u =", u1)
print("v =", v)
print("v1 =", v1)
print("||v1|| =", np.linalg.norm(v1))
print("redondeado al entero más cercano =", round(np.linalg.norm(v1)))

# Producto externo: u u^T como matriz
print("u u^T =")
print(np.outer(u, u))
```

## 7. Puente hacia QBronze: vectores de probabilidad

Antes de un qubit, QBronze usa estados clásicos probabilísticos. Un bit clásico probabilístico puede representarse con un vector:

$$p = \begin{bmatrix}p(0) \\ p(1)\end{bmatrix}, \quad p(0)+p(1)=1.$$

Ejemplo: una moneda justa tiene estado $[0.5, 0.5]$.

```python
p = np.array([0.5, 0.5])
print("estado probabilístico =", p)
print("suma de probabilidades =", p.sum())

# Operador clásico que intercambia 0 y 1.
X_clasico = np.array([[0, 1],
                      [1, 0]])

print("X_clasico @ p =", X_clasico @ p)
```
