```python
# Pregunta 1
"""
wich of the following vector are valid quantum States
"""

import numpy as np

vectores = {
    "vector 1": np.array([1/np.sqrt(2), -1/np.sqrt(2), 0, 0]),
    "vector 2": np.array([ 1/2, 0, 1/2, 0]),
    "vector 3": np.array([ 1/np.sqrt(2), -1/np.sqrt(2), -1/np.sqrt(2), -1/np.sqrt(2)]),
    "Vector 4": np.array([2/3, 1/3, 2/3, 0]),
    "Vector 5": np.array([1/2, -1/2, -1/2, 1/2]),
    "Vector 6": np.array([1/2, 1/2, 1/2, 1/2])
}

print("-- validacion de estados Cuanticos --")
for nombre, vec in vectores.items():
    norma_cuadrada = np.sum(np.abs(vec)**2)

    es_valido = np.isclose(norma_cuadrada, 1.0)

    estado= "Valido" if es_valido else "invalido"
    print(f"{nombre}: Suma de cuadrados = {norma_cuadrada:4f} -> {estado}")
```

### pregunta 2

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# 1. Creamos un circuito con 1 qubit
qc = QuantumCircuit(1)

# 2. El qubit por defecto inicia en |0>. 
# Aplicamos una compuerta X (NOT) para ponerlo en el estado |1>
qc.x(0)

# 3. Aplicamos la compuerta de Hadamard (H) al qubit 0
qc.h(0)

# 4. Obtenemos el vector de estado resultante
estado_final = Statevector.from_instruction(qc)

print("--- Resultado de aplicar H a |1> ---")
print("Vector de estado:")
print(estado_final.data)

# Mostramos el estado en notación de Dirac (ket)
print("\nNotación de Dirac:")
print(estado_final.draw())
```

```python
# 1. Importaciones necesarias de Qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 2. Definición del circuito cuántico
q = QuantumRegister(1, name="q")
c = ClassicalRegister(1, name="c")
qc = QuantumCircuit(q, c)

# Estado inicial por defecto: |0⟩
# Aplicamos la compuerta NOT (X) para cambiarlo al estado |1⟩
qc.x(q[0])

# Medimos el qubit y guardamos el resultado en el bit clásico
qc.measure(q[0], c[0])

# 3. Ejecución de la simulación
simulador = AerSimulator()
job = simulador.run(qc, shots=1024)
resultados = job.result()
counts = resultados.get_counts(qc)

# 4. Salidas visuales optimizadas para Jupyter
print("Conteo de mediciones:", counts)
```
