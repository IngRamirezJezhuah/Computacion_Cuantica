```python
import cirq
import numpy as np

# 1. Crear el circuito y el registro cuántico de 3 qubits
circuit = cirq.Circuit()
qq = cirq.LineQubit.range(3)

# 2. Inicialización: Colocar todos los qubits en superposición uniforme
circuit.append(cirq.H.on_each(*qq))

# 3. El Oráculo: Reemplazamos 'a' con 0
# Esto crea una compuerta CZ donde qq[0] es el control y qq[2] es el objetivo.
CCCZ = cirq.Z(qq[2]).controlled_by(qq[0])
circuit.append(CCCZ)

# --- Impresión del circuito para análisis visual ---
print("--- Estructura del Circuito Cuántico ---")
print(circuit)
print("\n" + "="*50 + "\n")

# 4. Simulación para verificar el vector de estado resultante
simulator = cirq.Simulator()
result = simulator.simulate(circuit)

print("--- Análisis del Vector de Estado Resultante ---")
# Obtenemos las amplitudes de probabilidad de todos los estados base
state_vector = result.state_vector()

# Mapeamos cada amplitud con su correspondiente etiqueta de estado en binario
states_labels = ["000", "001", "010", "011", "100", "101", "110", "111"]

for i, amplitude in enumerate(state_vector):
    # Formateamos números complejos pequeños a valores legibles limpios
    real = np.round(amplitude.real, 4)
    imag = np.round(amplitude.imag, 4)
    comp_str = f"{real} + {imag}j" if imag != 0 else f"{real}"
    
    # Resaltamos los estados que el oráculo marcó
    if states_labels[i] in ["101", "111"]:
        print(f"Estado |{states_labels[i]}⟩: Amplitud = {comp_str}  <-- ¡MARCADO CON FASE NEGATIVA!")
    else:
        print(f"Estado |{states_labels[i]}⟩: Amplitud = {comp_str}")
```

```python
import cirq
from cirq import H, X, Z

# Inicialización
qq = cirq.LineQubit.range(4)
circuit = cirq.Circuit()
circuit.append(H.on_each(*qq))

# 1. Definición del Oráculo (Rellenado con [3])
def oracle():
    yield Z(qq[3]).controlled_by(*(qq[0:3]))

# 2. Definición del Operador de Difusión
def grover_diffusion():
    yield H.on_each(*qq)
    yield X.on_each(*qq)
    yield Z(qq[3]).controlled_by(*(qq[0:3]))
    yield X.on_each(*qq)
    yield H.on_each(*qq)

# 3. Bucle de Iteraciones (Rellenado con oracle() y grover_diffusion())
for i in range(3):
    circuit.append(oracle())
    circuit.append(grover_diffusion())

# Imprimir circuito resultante para comprobar validez
print(circuit)
```
